```python
# For AS15625
import matplotlib.pyplot as plt

import numpy as np

# Globally update the font size
plt.rcParams.update({'font.size': 18})
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

    
x = [1,2,3,4]
total_paths_y = [0, 47, 710, 8192]
valid_paths_y = [0, 17, 60, 85]

hundred_rov_y = [0, 8, 12,0]
zero_rov_y = [0, 9, 27, 4]
na_rov_y = [0, 0, 3, 20]
ninety_rov_y = [0, 0, 1, 1]


# plt.figure(figsize=(4, 3))  # Set the figure size to 4 inches wide and 3 inches tall

# Plot
plt.plot(x, valid_paths_y,'-o', label='valid')
plt.plot(x, na_rov_y, '-o', label='NA ROV')
plt.plot(x, hundred_rov_y, '-o', label='100% ROV')

plt.plot(x, zero_rov_y, '-o', label='0% ROV')

# plt.plot(x, ninety_rov_y, '-o',label='(90-99)% ROV') # For average ROV score
plt.plot(x, ninety_rov_y, '-o',label='~93% ROV')


plt.xlabel("No. of AS hops", fontsize=18)
plt.ylabel("No. of paths", fontsize=18)
# plt.title('No. of AS hops vs no. of paths for AS15625')
plt.xticks(x, x, fontsize=16)
# plt.yticks(y, x, fontsize=16)


plt.legend()
plt.legend().get_frame().set_facecolor('none')

#plt.legend(fontsize=16) # using a size in points
plt.savefig("/home/shyam/Downloads/AS15625_2.pdf", bbox_inches='tight')

plt.show()
```
